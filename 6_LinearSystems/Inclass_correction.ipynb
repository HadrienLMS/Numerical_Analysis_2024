```python

A = np.array([[1.,3.],[1.,-1]])
print("A=",A)


def qr_decomposition(Mat):
    n = A.shape[0]
    Q = np.zeros_like(Mat)
    R = np.zeros_like(Mat)
    for i in range(n):
        print("--------\ni=",i)
        # original vector
        ai = Mat[:,i]
        # initialize
        ai_90 = ai
        # remove projections from previous vectors
        for j in range(i):
            print("j=",j)
            qj = Q[:,j]
            ai_90 -= np.dot(ai,qj)*qj
            R[j,i] = np.dot(ai,qj)
        
        print("ai_90 = ", ai_90)        
        ai_90_norm = np.linalg.norm(ai_90)
        qi = ai_90/ai_90_norm
        print("qi = ", qi)
        R[i,i] = ai_90_norm
        Q[:,i] = qi
    
    return Q, R

Q,R = qr_decomposition(A)
print("Q=",Q)
print("R=",R)

print("\n",1/np.sqrt(2))
            
        
        
```
