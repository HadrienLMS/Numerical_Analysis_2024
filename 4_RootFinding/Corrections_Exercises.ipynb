## Exercise 2 - 3 - 4 - root finding in 1D


```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt
import math
import statistics

# Define the test functions and their derivatives
def f1(x):
    return x**2 - 4 * np.sin(x)

def df1(x):
    return 2 * x - 4 * np.cos(x)

def f2(x):
    return x**3 - 2 * x - 5

def df2(x):
    return 3 * x**2 - 2

def f3(x):
    return x * np.sin(x) - 1

def df3(x):
    return np.sin(x) + x * np.cos(x)

def f4(x):
    return np.exp(-x) - x

def df4(x):
    return -np.exp(-x) - 1

def f5(x):
    return x**3 - 3 * x**2 + 3 * x - 1

def df5(x):
    return 3 * x**2 - 6 * x + 3

def get_iteration_count(history):
    return len(history)

def plot_function(func,a,b,root):
    plt.figure(figsize=(6, 4))
    x = np.linspace(a,b, 1000)
    y = func(x)
    plt.plot(x, y)
    plt.axhline(0, color="black", linestyle="-", linewidth=2)  # x-axis
    plt.axvline(0, color="black", linestyle="-", linewidth=2)  # x-axis
    #plt.set_title(label, fontsize=12)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    #plt.legend()
    plt.plot(root,0,'o')
    plt.tight_layout()
    plt.show()

# Bisection method
def bisection(f, a, b, tol=1e-8, max_iter=100):
    assert f(a) * f(b) < 0, "f(a) and f(b) must have opposite signs"
    history = []
    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        history.append(c)
        if abs(fc) < tol or (b - a) / 2 < tol:
            return c, history
        if f(a) * fc < 0:
            b = c
        else:
            a = c
    raise RuntimeError("Bisection method did not converge")

def convergence_rate(history,root):
    rates = []
    errors = []
    #print(history)
    for n in range(len(history)):  # Start from n=2 for enough data points
        errors.append(np.abs(history[n] - root)) 
    
    for n in range(2, len(history)):  # Start from n=2 for enough data points
        rate = np.log(errors[n] / errors[n-1]) / np.log(errors[n-1] / errors[n-2])
        rates.append(rate)
    
    # Filter out the negative numbers
    positive_rates = [x for x in rates if x > 0]
    #print(positive_rates)
    return statistics.mean(positive_rates)
    
# Newton's method
def newton(f, df, x0, tol=1e-8, max_iter=100):
    history = [x0]
    x = x0
    for _ in range(max_iter):
        fx = f(x)
        dfx = df(x)
        if abs(fx) < tol:
            return x, history
        if dfx == 0:
            raise RuntimeError("Zero derivative encountered")
        x_new = x - fx / dfx
        history.append(x_new)
        if abs(x_new - x) < tol:
            return x_new, history
        x = x_new
    raise RuntimeError("Newton's method did not converge")

# Secant method - start with two inputs x0 and x1
def secant(f, x0, x1, tol=1e-8, max_iter=100):
    history = [x0, x1]
    for _ in range(max_iter):
        fx0, fx1 = f(x0), f(x1)
        if abs(fx1) < tol:
            return x1, history
        if x1 == x0:
            raise RuntimeError("Division by zero in secant method")
        x_new = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
        history.append(x_new)
        if abs(x_new - x1) < tol:
            return x_new, history
        x0, x1 = x1, x_new
    raise RuntimeError("Secant method did not converge")

# Test functions
# define a list of tuples that contains all information
test_cases = [
    (f1, df1, 0.5, 3),
    (f2, df2, 0, 3),
    (f3, df3, -0.5, 2),
    (f4, df4, 0, 1),
    (f5, df5, -1, 2)
]

# Perform tests
for i, (f, df, a, b) in enumerate(test_cases, start=1):
    print(f"\nTesting function f{i}...")
    
    print("built-in root scalar method")
    builtin_sol = root_scalar(f, bracket=[a, b], method='brentq',xtol=1e-12, rtol=1e-12)
    print(f"  Root: {builtin_sol.root}, Iterations: {builtin_sol.iterations}")
    
    print("Bisection method:")
    root_bisection, history_bisection = bisection(f, a, b)
    rate_bisection = convergence_rate(history_bisection,builtin_sol.root)    
    #print(history_bisection)
    print(f"  Root: {root_bisection}, Iterations: {len(history_bisection)}, avg rate: {rate_bisection}")

    print("Newton's method:")
    try:
        root_newton, history_newton = newton(f, df, (a + b) / 2)
        rate_newton = convergence_rate(history_newton,builtin_sol.root)
        print(f"  Root: {root_newton}, Iterations: {len(history_newton)}, avg rate: {rate_newton}")
    except RuntimeError as e:
        print(f"  {e}")
        
    print("Secant method:")
    try:
        root_secant, history_secant = secant(f, a, b)
        rate_secant = convergence_rate(history_secant,builtin_sol.root)
        print(f"  Root: {root_secant}, Iterations: {len(history_secant)}, avg rate: {rate_secant}")
    except RuntimeError as e:
        print(f"  {e}")
        
    plot_function(f,a,b,root_newton)

        
```


## Exercise 3

For the four ﬁxed-point problems, we have
1. $g′(x) = 2x$, so $g′(2) = 4$, and hence ﬁxed-point iteration diverges.
2. $g′(x) = 1/(2 \sqrt{x + 2})$, so $g′(2) = 1/4$, and hence ﬁxed-point iteration converges
linearly with $C = 1/4$. The positive sign of $g′(2)$ causes the iterates to approach the ﬁxed-point from one side.
3. $g′(x) = −2/x^2$, so $g′(2) = −1/2$, and hence ﬁxed-point iteration converges linearly with $C = 1/2$. The negative sign of $g′(2)$ causes the iterates to spiral around the ﬁxed-point, alternating sides.
4. $g′(x) = (2x^2 − 2x − 4)/(2x − 1)^2$, so $g′(2) = 0$, and hence ﬁxed-point iteration converges quadratically


```python
def fixed_point(g,x0,max_iter,tol):
    x_new = x0
    n_steps = 1 # number of iterations
    err = 1 # initialize error
    while (err > tol) and (n_steps < max_iter):
        x = x_new
        x_new = g(x)
        err = np.abs(x - x_new) 
        n_steps+=1
        if (err > 10**10):
            raise ValueError("algorithm diverges !")
    print ("Fixed point method stopped after %i" %n_steps, "iterations, final error: %s" %err)
    return x

# for fixed point method
g = lambda x: x**2 - 2
g = lambda x: np.sqrt(x + 2)
g = lambda x:  1 + 2/x
g = lambda x:  (x**2 + 2)/(2*x-1)

It_max = 100; tol = 1e-10
x1 = fixed_point(g, 8, It_max, tol)
print("Method 1: found fixed-point %s:" %x1)

z = np.linspace(0.1,3,200)

plt.figure(figsize = (5, 3))
ax = plt.gca()
plt.plot(z, g(z), z, z)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
ax.set_xlim([0,3])
ax.set_ylim([0,3])


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define Newton's method (same as before)
def newton_2d_system(f, J, x0, tol=1e-12, max_iter=100):
    x = x0
    for i in range(max_iter):
        Fx = f(x)
        Jx = J(x)
        try:
            delta = np.linalg.solve(Jx, -Fx)
        except np.linalg.LinAlgError:
            print("Jacobian is singular. Stopping.")
            return x, False
        x = x + delta
        if np.linalg.norm(delta) < tol:
            return x, True
    print("Newton's method did not converge.")
    return x, False


# Define System 1
def f1(x):
    return np.array([
        x[0] * x[1] + x[1]**2 - 1,
        x[0] * x[1]**3 + x[0]**2 * x[1]**2 + 1
    ])

def J1(x):
    return np.array([
        [x[1], x[0] + 2 * x[1]],
        [x[1]**3 + 2 * x[0] * x[1]**2, 3 * x[0] * x[1]**2 + 2 * x[0]**2 * x[1]]
    ])

## Define System 2
#def f2(x):
#    return np.array([
#        x[0]**2 - x[1] + x[0] * np.cos(np.pi * x[0]),
#        x[1] * x[0] - 1 / x[0] + np.exp(-x[1])
#    ])

#def J2(x):
#    return np.array([
#        [2 * x[0] + np.cos(np.pi * x[0]) - np.pi * x[0] * np.sin(np.pi * x[0]), -1],
#        [x[1] + 1 / (x[0]**2), x[0] - np.exp(-x[1])]
#    ])


# Plotting function
def plot_system(f, x_sol, x_range, y_range, title):
    x = np.linspace(*x_range, 500)
    y = np.linspace(*y_range, 500)
    X, Y = np.meshgrid(x, y)
    
    # Evaluate each equation of the system on the grid
    Z1 = np.zeros_like(X)
    Z2 = np.zeros_like(X)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z = f([X[i, j], Y[i, j]])
            Z1[i, j] = Z[0]
            Z2[i, j] = Z[1]
    
    # Create subplots
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # Plot the first equation
    contour1 = axes[0].contourf(X, Y, Z1, levels=20, cmap='RdYlBu', alpha=0.8)
    #contour2 = plt.contour(X, Y, Z2, levels=20, cmap='RdYlBu', alpha=0.7, linestyles='dashed')
    cbar1 = fig.colorbar(contour1, ax=axes[0])
    cbar1.set_label("Value of f1(x, y)")
    axes[0].scatter(*x_sol, color='red', label='Solution', zorder=5)
    axes[0].set_title("Contour of f1(x, y)")
    axes[0].set_xlabel("x")
    axes[0].set_ylabel("y")
    axes[0].legend()
    axes[0].grid()
    
    # Plot the second equation
    contour2 = axes[1].contourf(X, Y, Z2, levels=20, cmap='RdYlBu', alpha=0.8)
    cbar2 = fig.colorbar(contour2, ax=axes[1])
    cbar2.set_label("Value of f2(x, y)")
    axes[1].scatter(*x_sol, color='red', label='Solution', zorder=5)
    axes[1].set_title("Contour of f2(x, y)")
    axes[1].set_xlabel("x")
    axes[1].set_ylabel("y")
    axes[1].legend()
    axes[1].grid()
    
    plt.tight_layout()
    plt.show()
    

# Solve and plot System 1
x0_1 = np.array([0.5, 0.5])
solution_1, success_1 = newton_2d_system(f1, J1, x0_1)
if success_1:
    print(f"Solution to System 1: {solution_1}")
    plot_system(f1, solution_1, (-2, 2), (-2, 2), "System 1: xy + y^2 = 1 and xy^3 + x^2y^2 = -1")

# Solve and plot System 2
#x0_2 = np.array([0.5, 0.5])
#solution_2, success_2 = newton_2d_system(f2, J2, x0_2)
#if success_2:
#    print(f"Solution to System 2: {solution_2}")
#    plot_system(f2, solution_2, (-2, 2), (-2, 2), "System 2: x^2 = y - x cos(πx) and yx = 1/x - e^(-y)")
```
