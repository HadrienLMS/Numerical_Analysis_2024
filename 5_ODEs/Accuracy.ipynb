## Exercise 1 - Truncation error 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of lambda * k values
lambda_h = np.linspace(-5, 5, 500)

# Exact solution
r_exact = np.exp(lambda_h)

# Forward Euler scheme
r_forward = 1 + lambda_h

# Backward Euler scheme
r_backward = 1 / (1 - lambda_h)

# Trapezoidal scheme
r_trapezoidal = (1 + lambda_h / 2) / (1 - lambda_h / 2)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(lambda_h, r_exact, label="Exact: $e^{\lambda h}$", linestyle='--', color='black')
plt.plot(lambda_h, r_forward, label="Forward Euler", color='blue')
plt.plot(lambda_h, r_backward, label="Backward Euler", color='orange')
plt.plot(lambda_h, r_trapezoidal, label="Trapezoidal", color='green')

# Formatting the plot
plt.axhline(1, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.title("Comparison of Multipliers ($r = U^{n+1}/U^n$)")
plt.xlabel("$\lambda h$")
plt.ylabel("$r$")
plt.legend()
plt.grid()

plt.xlim([-4, 1])
plt.ylim([-1, 3])

plt.axhline(0, color='black', linestyle='-', linewidth=2)  # Bold horizontal axis
plt.axvline(0, color='black', linestyle='-', linewidth=2)  # Bold vertical axis
```
