## Archimedes quadrature of the circle

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot a polygon
def plot_polygon(ax, radius, num_sides, color, label):
    angles = np.linspace(0, 2 * np.pi, num_sides, endpoint=False)
    x = radius * np.cos(angles)
    y = radius * np.sin(angles)
    ax.plot(np.append(x, x[0]), np.append(y, y[0]), color=color, label=label)

# Parameters
radius = 1
circle_theta = np.linspace(0, 2 * np.pi, 100)

# Set up plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.plot(radius * np.cos(circle_theta), radius * np.sin(circle_theta), color='black', label="Circle")
ax.set_title("Archimedes' Quadrature of the Circle")

# Plot polygons with increasing number of sides
sides_list = [6, 12, 24, 96]  # Different levels of approximation
colors = ['blue', 'green', 'orange', 'red']

for i, sides in enumerate(sides_list):
    plot_polygon(ax, radius, sides, color=colors[i], label=f'{sides}-sided polygon')

# Plot settings
ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

## Legendre polynomials - roots

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import roots_legendre

def Legendre(x,k):
    if (k == 0):
        return 1
    elif (k == 1):
        return x
    else:
        return  (1 / k) * ( (2*(k-1)+1)*x*Legendre(x,k-1) - (k-1)*Legendre(x,k-2) )

x = np.linspace(-1, 1, 100)
plt.figure(figsize = (12, 8))
n = 5
for i in range(1,n+1):
    plt.plot(x, Legendre(x,i), label="L_%i" %i)
    plt.plot(roots_legendre(i)[0], np.zeros(i), "k*")
plt.title("First Legendre polynomials")
plt.xlabel("x")
plt.legend();
```
